```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
```

```python
x=np.linspace(0,5,11) #nos va a dar once elementos del 0 al 5 en forma de array
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) #nos traza la gráfica de la función que hicimos arriba, ahora bien veamos con la raíz cuadrada
```

```python
z = np.sqrt(x)
```

```python
plt.plot(x,z) #notemos que nos da el valor de salida, pero ahora quitemos el out de abajo, es decir, sólo imprimir la gráfica
plt.show()
```

```python
#ahora pongámosle un titulo a nuestra gráfica
plt.plot(x,z)
#plt.plot(x,y) esto si queremos poner dos gráficas en un sólo plano :)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
#veamos la manera de imprimir varias gráficas en esta celda
plt.subplot(2,1,1) #el último número representa el número de gráfica que es, o sea el orden
plt.plot(x,y)

plt.subplot(2,1,2) #enotemos que los primeros dos números nos indican cómo se pueden organizar los plot
plt.plot(x,z)
```

```python
# OO
fig = plt.figure() #hemos creado un objeto figura 

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #los primeros dos argumentos se refieren a la numeración de x y y, los otros dos al tamaño de largo y ancho
#notemos quelo de aquí arriba es la tabla o lienzo donde dibujamos la gráfica
axes.plot(x,y)
axes.plot(x,z)
#ahora veamos si les ponemos nombres a los ejes

axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5]) #notemos que el primer argumento se refiera a qué tanto se moverá horizontalmetnte, eje x y el segundo habla sobre qué tanto se moverá verticalmente, eje y
#los últimos argumentos ya sabemos qye se refiere a ancho y altura respectivamente 
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
for current in axes: 
   current.plot(x,y)
#veamos que también podemos indexar las gráficas es decir:
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel("X label of axe 0")

axes[1].plot(x,z)
axes[1].set_xlabel("X label of axe 1")

plt.tight_layout() #es recomendable poner esto al final de cada ejercicio, dado que pueden quedar muy estrechos o no
```

```python
axes
```

```python
fig = plt.figure(figsize=(5,2)) #nótese que el primer argumento se refiere al largo y el segundo al ancho 
#podrías hacer lo mismo de arriba con subplots
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,5))
axes[0].plot(x,y)
axes[1].plot(x,z)

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(5,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label="f(x)=x²")
ax.plot(x,z,label="f(x)=sqrt(x)")
#agregamos el label en la descripción pq queremos ponerle etiquetas a las curvas trazadas
ax.legend()
#se recomienda ver la pagina de matplotlib para poder saber qué parametros admite ax.legend()
```

```python
fig = plt.figure(figsize=(5,2))
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color="red",linewidth=3,alpha=0.5) #aquí especificamos el color al igual que la longitud de la línea 
#alpha se refiere a q tantlo vamos a poner "clarita".veamos otra variedad del estilo de la linea
ax.plot(x,y,color="red",lw=3,linestyle="--")
ax.plot(x,z,color="yellow",linewidth=3, linestyle=":")
```
